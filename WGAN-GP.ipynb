```python
import torch
import torch.nn as nn
import torch.optim as optim
from torchvision import datasets, transforms
from torch.utils.data import DataLoader
import torchvision
import matplotlib.pyplot as plt
import numpy as np

# Data preprocessing
transformations = transforms.Compose([   #,transforms.Normalize((0.5,), (0.5,))
    
    transforms.ToTensor()
])

data = datasets.FashionMNIST(
    root="data",
    train=True,
    download=True,
    transform=transformations
)


class Critic(nn.Module): #in wasserstein GAN the Discriminator does not have a binary output (no activation function at the end of the network) so its rather a Critic 
    def __init__(self, img_dim):
        super(Critic, self).__init__()
        self.fc1 = nn.Linear(img_dim, 128)
        self.activation1 = nn.LeakyReLU(0.1)
        self.fc2 = nn.Linear(128, 64)
        self.activation2 = nn.LeakyReLU(0.1)
        self.fc3 = nn.Linear(64, 1)


    def forward(self, x):
        x = self.fc1(x)
        x = self.activation1(x)
        x = self.fc2(x)
        x = self.activation2(x)
        x = self.fc3(x)  
        return x

class Generator(nn.Module):
    def __init__(self, z_dim, img_dim):
        super(Generator, self).__init__()
        self.fc1 = nn.Linear(z_dim, 256)
        self.bn1 = nn.BatchNorm1d(256)
        self.activation1 = nn.LeakyReLU(0.1)
        self.fc2 = nn.Linear(256, img_dim) 
        self.output_activation = nn.Sigmoid()

    def forward(self, x):
        x = self.fc1(x)
        x = self.activation1(x)
        x = self.bn1(x)
        x = self.fc2(x)
        x = self.output_activation(x)
        return x


def gradient_penalty(critic, real, fake):
    batch_size = real.size(0)
    # Generate random epsilon
    epsilon = torch.rand(batch_size, 1, device=real.device)
    # Interpolate between real and fake data
    interpolated = epsilon * real + (1 - epsilon) * fake
    interpolated.requires_grad_(True)

    # Calculate critic scores for interpolated images
    critic_interpolated = critic(interpolated)

    # Calculate gradients of scores with respect to the interpolated images
    gradients = torch.autograd.grad(
        outputs=critic_interpolated,
        inputs=interpolated,
        grad_outputs=torch.ones_like(critic_interpolated),
        create_graph=True,
        retain_graph=True,
        only_inputs=True,
    )[0]

    # Calculate the L2 norm of the gradients
    gradients = gradients.view(batch_size, -1)
    gradient_norm = gradients.norm(2, dim=1)
    gradient_penalty = ((gradient_norm - 1) ** 2).mean()
    return gradient_penalty

# Hyperparameters
lr = 0.00005  # Learning rate 
z_dim = 100 # dimensionality of latent vector space 
img_dim = 28*28  # 784 for FashionMNIST
batch_size = 32 
num_epochs = 201

critic_iters = 5  # The number of critic updates per generator update
lambda_gp = 10



# Models
critic = Critic(img_dim)
generator = Generator(z_dim, img_dim)

# Optimizers (Using RMSprop as recommended for WGAN)
optim_critic = optim.RMSprop(critic.parameters(), lr=lr)
optim_generator = optim.RMSprop(generator.parameters(), lr=lr)

loader = DataLoader(data, batch_size=batch_size, shuffle=True)

# Training Loop
for epoch in range(num_epochs):
    for batch_idx, (real, _) in enumerate(loader):
        real = real.view(-1, 784)

        # Train Critic
        for _ in range(critic_iters):
            noise = torch.randn(batch_size, z_dim)
            fake = generator(noise)
            critic_real = critic(real).reshape(-1)
            critic_fake = critic(fake).reshape(-1)
            loss_critic = -(torch.mean(critic_real) - torch.mean(critic_fake)) #Since we want to minimize the loss of the critic we maximize the negative 

            gp_value = gradient_penalty(critic, real.data, fake.data)
            loss_critic += lambda_gp * gp_value

            optim_critic.zero_grad()
            loss_critic.backward()
            optim_critic.step()

        # Train Generator
        noise = torch.randn(batch_size, z_dim)
        fake = generator(noise)
        output = critic(fake).reshape(-1)
        loss_gen = -torch.mean(output)
        optim_generator.zero_grad()
        loss_gen.backward()
        optim_generator.step()

        # Print losses at the start of the epoch
        if batch_idx == 0:
            print(f"Epoch [{epoch+1}/{num_epochs}] \
                  Loss Critic: {loss_critic:.4f}, Loss Gen: {loss_gen:.4f}")

        
        if epoch in (0,25,50,75,100,125,150,175,200) and batch_idx == 0:
            with torch.no_grad():
                fixed_noise = torch.randn(32, z_dim)  # Assuming you want to generate 32 images
                fake_images = generator(fixed_noise).reshape(-1, 1, 28, 28)
                img_grid_fake = torchvision.utils.make_grid(fake_images, normalize=True)

                # Convert image grid to numpy and plot
                img_np = img_grid_fake.permute(1, 2, 0).numpy()
                plt.figure(figsize=(8, 8))
                plt.imshow(np.clip(img_np, 0, 1))
                plt.title(f"Epoch {epoch}, Batch {batch_idx}")
                plt.axis('off')
                plt.show()
```
